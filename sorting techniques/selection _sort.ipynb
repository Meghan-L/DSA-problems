```python
def selection_sort(arr):
    for i in range(len(arr)):
        least=i
        for j in range(i,len(arr)):
            if arr[least]>arr[j]:
                least=j
        arr[i],arr[least]=arr[least],arr[i]
    return arr
```

```python
selection_sort([13, 46, 24, 52, 20, 9])
```
