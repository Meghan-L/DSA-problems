```python
def union_of_arr(arr1,arr2):
    freq={}
    for i in arr1:
        freq[i]=1
    for i in arr2:
        if freq.get(i,0)==0:
            freq[i]=1
            arr1.append(i)
    return arr1
```

algorithm - time O(n) space (n+m) <br>
step 1 create a hash map of all existing elements in arr1 <br>
step 2 start iterating in the second array arr2 <br>
step 3 check if there is a value in arr2 not present in arr1 <br>
step 4 append that value to first array <br>

```python
def union_arr(arr1,arr2):
    arr1=set(arr1)
    arr2=set(arr2)
    for i in arr2 :
        if i not in arr1:
            arr1.add(i)
    return arr1 

def simple_union(arr1,arr2):
    arr1,arr2=set(arr1),set(arr2)
    return arr1 | arr2
```
