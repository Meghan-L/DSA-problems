algorithm - time O(n) space O(N) <br>
step 1 create a hash map of number of occuerence of each element <br>
step 2 while counting check if the count>N//2 <br>
step 3 return that particular value 

```python
# moores voting algorithm 
def occurence_element(arr):
    element=arr[0]
    count=1
    for i in range(1,len(arr)):
        if count==0:
            element=arr[i]
        if arr[i]==element:
            count+=1
        else:
            count-=1
    if arr.count(element) > len(arr)//2:
        return element
    return -1 
```
