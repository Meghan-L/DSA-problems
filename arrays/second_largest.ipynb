```python
#brute force 
def second_largest(arr):
    arr.sort()
    return arr[-2]
```

algorithm - time O(nlogn) and space O(1)<br>
step 1 sort the array in ascending order <br>
step 2 return the second element from last <br>
Note you can do the same for minimum also just return second element from first

```python
def second_biggest(arr):
    if len(arr)<2:
        return None
    second=first=float("-inf")
    for i in range(len(arr)):
        if arr[i]>first :
            first,second=arr[i],first
        elif first>arr[i]>second:
            second=arr[i]
    return second
```

algorithm - time O(n) space O(1)<br>
step 1 return none if the number of elements is less than 2<br>
step 2 initialise the first and second largest as the minimum number possible <br>
step 3 starting iterating over the list <br>
step 4 if a number is greater than first swap the values of first with that number and value of first into second <br>
step 5 if a number is greater than second and less than first update second <br>
Note you can do the same for second minimum by adding more if conditions
