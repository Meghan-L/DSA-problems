```python
#brute force
def duplicate_elements(arr):
    arr.sort()
    for i in range(1,len(arr)):
        if arr[i-1]==arr[i]:
            return True
    return False
```

algorithm - time O(nlogN) space(1) <br>
step 1 sort the array in ascending order <br>
step 2 duplicate values will be side by side <br>
step 3 check each and element side by side 

```python
def duplicate_values(arr):
    freq={}
    for i in arr:
        freq[i]=freq.get(i,0)+1
    for i in freq.values():
        if i>1:
            return True
    return True 
```

algorithm - time O(N) space O(N)<br>
step 1 create a hash map containing the number of times the value occured <br>
step 2 iterate over the values in that hash map <br>
step 3 if there is a value that is greater than 1 return True <br>
