```python
#brute force method 
def stock_buy(arr):
    n=len(arr)
    profit=0
    for i in range(n):
        for j in range(i+1,n):
            if arr[j]-arr[i]>profit:
                profit=arr[j]-arr[i]
    return profit 
```

algorithm - time O(n^2) space O(1)<br>
step 1 select the value at first index <br>
step 2 check difference between the values other than first index <br>
step 3 if any value is greater than profit change profit <br>
step 4 repeat above process for all the elements in the array <br>

```python
def stock_buy(arr):
    min_val=arr[0]
    max_val=0
    for i in arr:
        max_val=max(max_val,i-min_val)
        min_val=min(min_val,i)
    return max_val
```

algorithm - time O(n) space O(1) <br>
step 1 find the minimum value uptil that point <br>
step 2 find the maximum positive difference till that point  which is curr_val-minimum value<br>
step 3 return the maximum difference <br>
