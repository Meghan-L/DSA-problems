```python
def problem1(arr):
    freq={}
    for i in arr:
        freq[i]=freq.get(i,0)+1
    for i in freq:
        if freq[i]==1:
            return i
```

```python
def problem2(arr):
    arr.sort()
    for i in range(1,len(arr),2):
        if arr[i-1]==arr[i]:
            continue
        else:
            return arr[i-1]
    return arr[-1]
```
