```python
def consecutive_ones(arr):
    count=result=0
    for i in arr:
        if i == 1:
            count+=1
            if count>result:
                result=count
        else:
            count=0
    return result
```

algorithm - time O(N) space O(1) <br>
step 1 start iterating through the array <br>
step 2 if a value is 1 then increment count <br>
step 3 if consecutive value is not also 1 then we set count to zero <br>
step 4 we return the maximum the count we can get
