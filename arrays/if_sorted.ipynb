```python
def sorted_arr(arr):
    return arr==arr.sort()
```

algorithm - time O(nlogn) space O(1)<br>
step 1 compare the arr with the sorted form of arr <br>
step 2 return the compared value <br>

```python
def is_sorted(arr):
    for i in range(1,len(arr)):
        if arr[i-1]>arr[i]:
            return False
    return True 
```

algorithm - time O(N) spaceO(1) <br>
step 1 compare adjacent values <br>
step 2 if the next value is less than current value return false <br>
