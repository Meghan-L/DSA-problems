```python
def sort_color(arr):
    color={0:0,1:0,2:0}
    k=0
    for i in arr:
        color[i]+=1
    for i in [0,1,2]:
        while color[i]>0:
            arr[k]=i
            k+=1
            color[i]-=1
    return arr
```

```python
#dutch national flag 
def sort_color(arr):
    n=len(arr)
    low=mid=0
    high=n-1
    while mid<=high:
        if arr[mid]==0:
            arr[mid],arr[low]=arr[low],arr[mid]
            mid+=1
            low+=1
        if arr[mid]==1:
            mid+=1
        if arr[mid]==2:
            arr[high],arr[mid]=arr[mid],arr[high]
            high-=1
    return arr
```
