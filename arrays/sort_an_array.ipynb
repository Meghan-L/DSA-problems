```python
def sort_color(arr):
    color={0:0,1:0,2:0}
    k=0
    for i in arr:
        color[i]+=1
    for i in [0,1,2]:
        while color[i]>0:
            arr[k]=i
            k+=1
            color[i]-=1
    return arr
```
