```python
def sum_2(arr,target):
    n=len(arr)
    l=0
    r=n-1
    arr.sort()
    while l<r:
        if arr[l]+arr[r]==target:
            return [l,r]
        if arr[l]+arr[r]>target:
            r-=1
        if arr[l]+arr[r]<target:
            l+=1
    return []
```

```python
def sum_2(arr,target):
    freq={}
    for i in range(len(arr)):
        freq[arr[i]]=i
        if target - i not in freq:
            return [freq[target-i],i]
    return []
```
