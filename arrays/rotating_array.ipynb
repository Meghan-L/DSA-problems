```python
def rotate_array(arr,k):
    res=[]
    for i in range(k,len(arr)):
        res.append(arr[i])
    for i in range(k):
        res.append(arr[i])
    return res
```

algorithm - time O(N) space(N) <br>
step 1 initialise  a result array 
step 2 start iterating from the kth index to n-1<br>
step 3 add each and every element <br>
step 4 initialise a for loop to add values in the indexes 0 to k<br>
