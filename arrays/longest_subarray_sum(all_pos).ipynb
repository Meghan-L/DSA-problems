```python
def subarray_sum(arr,k):
    result=sum_arr=count=0
    for i in range(len(arr)):
        sum_arr=arr[i]
        if sum_arr==k:
            count+=1
        
        for j in range(i+1,len(arr)):
            sum_arr+=arr[j]
            if sum_arr==k:
                count+=1
            
    return count
```

algorithm - time O(n^2) space O(1) <br>
step 1 start iterating for every sub array <br>
step 2 find the sum of each element <br>
step 3 return the sub array with maximum sum <br>

```python
def long_subbaray_sum(arr,k):
    n=len(arr)
    point=0
    sum_arr=0
    result=count=0
    for i in range(n):
        sum_arr+=arr[i]
        count+=1
        if sum_arr==k:
            if count>result:
                result=count
        while sum_arr>k:
            sum_arr-=arr[point]
            point+=1
            count-=1 
            if sum_arr==k:
                if count>result:
                    result=count       
    return result
```

algorithm - time  O(n) space O(1)<br>
step 1 intialise a sliding window <br>
step 2 as we only need the sum increase the value of sum as each element is seen <br>
step 3 increase count as each element is added to sum <br>
step 4 if the sum is greater than the target remove elements present in left and decrement the count <br>

```python
def max_len_sub_array(arr,k):
    n=len(arr)
    prefix={}
    count=result=0
    sum_arr=0
    for i in range(n):
        sum_arr+=arr[i]
        if sum_arr==k:
            result=i+1
        if sum_arr-k in prefix:
            count=i-prefix[sum_arr-k]
            if count>result:
                result=count
        if sum_arr not in prefix:
            prefix[sum_arr]=i
    return result 
```
