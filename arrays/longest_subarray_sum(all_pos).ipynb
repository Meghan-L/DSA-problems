```python
def subarray_sum(arr,k):
    result=sum_arr=count=0
    for i in range(len(arr)):
        sum_arr=arr[i]
        if sum_arr==k:
            count+=1
        
        for j in range(i+1,len(arr)):
            sum_arr+=arr[j]
            if sum_arr==k:
                count+=1
            
    return count
```

```python
def long_subbaray_sum(arr,k):
    n=len(arr)
    point=0
    sum_arr=0
    result=count=0
    for i in range(n):
        sum_arr+=arr[i]
        count+=1
        if sum_arr==k:
            if count>result:
                result=count
        while sum_arr>k:
            sum_arr-=arr[point]
            point+=1
            count-=1 
            if sum_arr==k:
                if count>result:
                    result=count       
    return result
```

algorithm - time  O(n) space O(1)<br>
step 1 intialise a sliding window <br>
step 2 as we only need the sum increase the value of sum as each element is seen <br>
step 3 increase count as each element is added to sum <br>
step 4 if the sum is greater than the target remove elements present in left and decrement the count <br>

```python
def max_len_sub_array(arr,k):
    n=len(arr)
    prefix={}
    count=result=0
    sum_arr=0
    for i in range(n):
        sum_arr+=arr[i]
        if sum_arr==k:
            result=i+1
        if sum_arr-k in prefix:
            count=i-prefix[sum_arr-k]
            if count>result:
                result=count
        if sum_arr not in prefix:
            prefix[sum_arr]=i
    return result 
```

algorithm - time o(n) space O(1) <br>
step 1 as we reach every element add the number at that index to sum <br>
step 2 if the sum is not present already then store it in prefix with the index value in hash map <br>
step 3 if the sum is equal to the target we can change the result to i+1 <br>
step 4 if the target - number is already present in prefix then length = present index - index of [target-number]<br>

```python
# only for positive elements 
def long_subarray_sum(arr,k):
    n=len(arr)
    j=count=sum_arr=result=0
    for i in range(n):
        sum_arr+=arr[i]
        count+=1
        while sum_arr>k:
            sum_arr-=arr[j]
            j+=1
            count-=1
        if sum_arr==k:
            if result<count:
                result=count
    return result
```
