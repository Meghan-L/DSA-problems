```python
# brute force approach 
def large_element(arr):
    arr.sort()
    return arr[-1]
```

algorithm  - time O(nlogn) space O(1)<br>
step 1 sort the array in ascending order <br>
step 2 return the last element 

```python
# optimal 
def big_element(arr):
    max_ele=arr[0]
    for i in range(1,len(arr)):
        if arr[i]>max_ele:
            max_ele=arr[i]
    return max_ele
```

algorithm - time O(n) space O(1)<br>
step 1 initialise the first element as the max of the array <br>
step 2 using a for loop  to check each and every element <br>
step 3 if an element is greater than the current max chaange them
