```python
def kadanes_algo(arr):
    n=len(arr)
    sum_arr=res=arr[0]
    start=end=temp_start=0
    for i in range(1,n):
        sum_arr+=arr[i]
        
        if sum_arr <arr[i]:
            sum_arr=arr[i]
            temp_start=i
        if sum_arr>res:
            res=sum_arr
            start=temp_start
            end=i

    return arr[start:end+1]
```

algorithm - time O(n) space O(1) <br>
step 1 start adding the value as u see them in the array <br>
step 2 if the sum uptil now is less than the next index value change sum and start position <br>
step 3 if the sum obtained is maximum save the sum and starting and ending values <br>
step 4 return the sub array by slicing it with the help of start and end <br>
