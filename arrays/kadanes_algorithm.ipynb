```python
def kadanes_algo(arr):
    n=len(arr)
    sum_arr=result=0
    res=[]
    arr2=[]
    for i in range(n):
        sum_arr+=arr[i]
        arr2.append(arr[i])
        if sum_arr<0:
            arr2=[]
            sum_arr=0
        if sum_arr>result:
            result=sum_arr
            res=arr2
    return res
```
