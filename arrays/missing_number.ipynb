```python
def missing_number(arr):
    n=len(arr)
    check=[x for x in range(1,n)]
    for i in arr:
        if i not in check:
            return i
    
```

algorithm - time O(n) space O(n) <br>
step 1 create a check array consisting of all the values from 0 to n-1 <br>
step 2 iterate through every element in the given array <br>
step 3 check if the particular value not in check array return that value <br>
